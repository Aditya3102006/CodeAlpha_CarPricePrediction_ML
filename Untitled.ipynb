##### 3. Convert other categorical columns using get_dummies method in pandas

```python
df_encoded = pd.get_dummies(df, columns=['Manufacturer', 'Model', 'Category', 'Fuel type', 'Gear box type', 'Drive wheels', 'Wheel', 'Color'],
                           drop_first=True)

 # 1. pd.get_dummies(df, columns=[...])
 #     This tells Pandas:
 #     “Take these columns which have text values (like 'Petrol', 'Diesel', 'CNG'), and create new columns for them with only 0s and 1s.” 
 #     This is called One-Hot Encoding.
 #     Example:Let’s say this is your data:
 #     Fuel type
 #     Petrol
 #     Diesel
 #     Petrol
 #     By default, get_dummies() will create:
 #     Fuel_Diesel	Fuel_Petrol
 #     0	            1
 #     1	            0
 #     0	            1
 # 2. What does drop_first=True mean?
 #     It means:Drop the first dummy column to avoid multicollinearity in machine learning.
 #     So instead of two columns (Fuel_Diesel and Fuel_Petrol), it will keep only one column:
 #     Fuel_Diesel
 #     0
 #     1
 #     0
 #     You can still tell:If Fuel_Diesel is 1 → it’s diesel , If Fuel_Diesel is 0 → then it’s petrol
 #     -This saves space and avoids confusion for the model.

 # 3. What happens if you remove drop_first=True?
 #     df_encoded = pd.get_dummies(df, columns=[...])
 #     This will create all dummy columns (e.g., both Fuel_Petrol and Fuel_Diesel).
 #     It works fine — no error will happen — but:
 #     It adds one extra column per category
 #     It may cause multicollinearity in linear regression (model confusion due to duplicate info)
    
 #     So, even though you see no error or visible change, internally it’s better to use drop_first=True — especially with Linear Regression.
```
